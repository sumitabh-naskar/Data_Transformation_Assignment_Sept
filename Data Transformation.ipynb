```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, MinMaxScaler, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Creating a sample dataset based on the assignment requirements
data = {
    'Age': [25, 30, np.nan, 45, 35],
    'Salary': [50000, 60000, 75000, np.nan, 80000],
    'Hours_Worked_Per_Week': [40, np.nan, 45, 50, 38],
    'Performance_Score': [8, 7, np.nan, 9, 6],
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'Department': ['IT', 'HR', 'IT', 'Sales', 'HR'],
    'Work_Mode': ['Remote', 'Office', 'Hybrid', 'Remote', 'Office'],
    'Location': ['Mumbai', 'Delhi', 'Bangalore', 'Mumbai', 'Delhi'],
    'Projects_Completed': [5, 10, 12, 7, 8]
}

df = pd.DataFrame(data)
```

```python
#Q1 :- Handle Missing Values (Basic Cleaning)

df['Age'] = df['Age'].fillna(df['Age'].median())
df['Performance_Score'] = df['Performance_Score'].fillna(df['Performance_Score'].median())
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
df['Hours_Worked_Per_Week'] = df['Hours_Worked_Per_Week'].fillna(df['Hours_Worked_Per_Week'].mean())

print("Dataset after handling missing values:")
print(df)
```

```python
#Q2 :- Label Encoding

le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
df['Department'] = le.fit_transform(df['Department'])

print("\nUpdated Gender and Department columns:")
print(df[['Gender', 'Department']].head())
```

```python
#Q3 :- One-Hot Encoding

df_encoded = pd.get_dummies(df, columns=['Work_Mode', 'Location'])

print("\nDataset with One-Hot Encoding (New columns created):")
print(df_encoded.columns)
print(f"Total columns now: {len(df_encoded.columns)}")
```

```python
#Q4 :- Normalization (Min-Max Scaling)

mms = MinMaxScaler()
df_encoded[['Salary', 'Hours_Worked_Per_Week']] = mms.fit_transform(df_encoded[['Salary', 'Hours_Worked_Per_Week']])

print("\nNormalized Salary and Hours Worked:")
print(df_encoded[['Salary', 'Hours_Worked_Per_Week']].head())
```

```python
#Q5 :- Standardization (Scaling)

ss = StandardScaler()
df_encoded[['Age', 'Projects_Completed']] = ss.fit_transform(df_encoded[['Age', 'Projects_Completed']])

print("\nStandardized Age and Projects Completed:")
print(df_encoded[['Age', 'Projects_Completed']].head())
```

```python
#Q6 :- Compare Scaling Methods

compare_df = pd.DataFrame()
compare_df['Original_Salary'] = [50000, 60000, 75000, 66250, 80000] # Sample values
compare_df['MinMax_Scaled'] = mms.fit_transform(compare_df[['Original_Salary']])
compare_df['Standard_Scaled'] = ss.fit_transform(compare_df[['Original_Salary']])

print("\nScaling Methods Comparison (Salary):")
print(compare_df)
```

```python
#Q7 & Q8. 
#Build and Apply Preprocessing Pipeline

# Defining column groups
num_cols = ['Age', 'Salary', 'Hours_Worked_Per_Week', 'Projects_Completed']
cat_cols = ['Work_Mode', 'Location']

# Create Pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_cols),
        ('cat', OneHotEncoder(), cat_cols)
    ])

pipeline = Pipeline(steps=[('preprocessor', preprocessor)])

# Apply to original dataframe (re-loading original 'df' for clean run)
transformed_data = pipeline.fit_transform(df)

print("\nShape of final dataset:", transformed_data.shape)
```

```python
#Q9 :- Why is scaling important in Python?

#Answer :- Most Machine Learning algorithms (like K-NN, SVM, and Linear Regression) use mathematical distances to calculate results. 
#If one feature has a range of $0-1$ and another has a range of $0-100,000$, the larger number will dominate the model's learning process. 
#Scaling ensures all features contribute equally to the result.
```

```python
#Q10 :- Why do we convert categorical data into numerical form?

#Answer :- Mathematical models and equations cannot "read" text or labels like "IT" or "Remote." Since Machine Learning models are essentially 
#complex mathematical functions, all input data must be converted into numbers (integers or floats) so the algorithms can perform computations like matrix multiplication and gradient descent.
```
